```python
g0 = pd.read_csv('{}reconstruction/group0/full_first.csv'.format(p4), header = [0,1], index_col = [0])
g1 = pd.read_csv('{}reconstruction/group1/full_first.csv'.format(p4),header = [0,1], index_col = [0])
g2 = pd.read_csv('{}reconstruction/group2/full_first.csv'.format(p4),header = [0,1], index_col = [0])
all_i = pd.read_csv('{}tot_pop/full_first.csv'.format(p4),header = [0,1], index_col = [0])
```

```python
g2
```

```python
g0[[('Treated','p-val')]] = g0[[('Test-TC', 'pval')]]
g1[[('Treated','p-val')]] = g1[[('Test-TC', 'pval')]]
g2[[('Treated','p-val')]] = g2[[('Test-TC', 'pval')]]
all_i[[('Treated','p-val')]] = all_i[[('Test-TC', 'pval')]]
```

```python
g0 = g0['Treated']
g1 = g1['Treated']
g2 = g2['Treated']
all_i = all_i['Treated']
```

```python
g0['N=767 mu (pval)'] = g0['mean'].round(2).astype(str)+' ('+g0['p-val'].round(2).astype(str)+')'
g1['N=516 mu (pval)'] = g1['mean'].round(2).astype(str)+' ('+g1['p-val'].round(2).astype(str)+')'
g2['N=81 mu (pval)'] = g2['mean'].round(2).astype(str)+' ('+g2['p-val'].round(2).astype(str)+')'
all_i['N=1364 mu (pval)'] = all_i['mean'].round(2).astype(str)+' ('+all_i['p-val'].round(2).astype(str)+')'
```

```python
mat = pd.concat((all_i, g0, g1, g2), axis =1 ,keys = ['Canary Population', 'No Effect Sick Group',
                                                     'Positive Effect Healthy Group', 'Positive Effect Critical Group'])
```

```python
mat.drop(columns = ['std', 'count','mean','p-val'], axis = 1,level = 1, inplace = True)
```

```python
mat.index = ['Age', 'Total Cost', 'Geneder', 'Tobacco Use', 'Pressure', 'Obesity', 'Hypertension',
            'Hypothyroid', 'Disease Count', 'Acute Care 2', 'Acute Care 6', 
             'Inpatient Care 6', 'Emergency Visits 6', 'Line of Bussiness', 'Acute6-bin']
```

```python
mat.round(2)
```

```python
agglo = pd.read_csv('{}reconstruction/Acute-Care_ddlrank_rmst.csv'.format(p4),index_col = 0)
tree = pd.read_csv('{}reconstruction/Acute-Care_ddlrank_rmst.csv'.format(p4tr),index_col = 0)
no_clust = pd.read_csv('{}reconstruction/Acute-Care_ddlrank_rmst.csv'.format(p4A),index_col = 0)
```

```python
agglo.drop(columns = ['test_statistic', '-log2(p)'], inplace = True)
tree.drop(columns = ['test_statistic', '-log2(p)'], inplace = True)
no_clust.drop(columns = ['test_statistic', '-log2(p)'], inplace = True)
```

```python
agglo.index = [['Group 0', 'Group 1', 'Group 2']]
tree.index = [['Group 1', 'Group 2', 'Group 0']]
no_clust.index = [['Group 1', 'Group 0', 'Group 2']]
```

```python
agglo.T
```

```python
agglo = agglo.T
tree = tree.sort_index().T
no_clust = no_clust.sort_index().T

mat_all = pd.concat((agglo, tree, no_clust), axis = 1, keys = ['Agglomerative', 'Decision Tree', 'No Clustering'])
```

```python
mat_all.drop(index = 'C', inplace = True)
```

```python
mat_all.index = ['pval', 'rsmt-6', 'rmst-12', 'rmst-18', 'rmst-24']
```

```python
mat_all.round(2).T.to_csv('rmstv1.csv')
mat_all.round(2).to_csv('rmstv2.csv')
```
