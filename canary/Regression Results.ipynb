```python
import pandas as pd
import numpy as np
from pathlib import Path
from IPython.display import display
```

```python
p = Path('results/whole_pop/effects/tables/survival')

def display2(p, name, flag = False):
    if "check" in p.name:
        return
    if p.is_dir():
        for l in p.iterdir():
            display2(l,p.name, flag)
    else:
        if flag:
            print(p.name)
        else:
            print(name)
        display(pd.read_csv(p, index_col = 0))
        
display2(p,"")
```
